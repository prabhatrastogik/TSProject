```python
from io import StringIO

from bs4 import BeautifulSoup
from requests import get

import pandas as pd
```

```python
BASE_URL = 'https://www1.nseindia.com'
HISTORIC_ENDPOINT = '/products/dynaContent/common/productsSymbolMapping.jsp'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) ' + \
            'AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.0.3 Safari/605.1.15',
}

query_params = {
    'symbol': 'IIFL',
    'segmentLink': '3',
    'symbolCount': '2',
    'series': 'EQ',
    'dateRange': ' ',
    'fromDate': '01-04-2020',
    'toDate': '01-04-2021',
    'dataType': 'PRICEVOLUMEDELIVERABLE',
}
```

```python
r = get(BASE_URL + HISTORIC_ENDPOINT, headers = headers, params=query_params)
r.status_code
```

```python
filename = "test.txt"
with open(filename, 'wb') as fd:
    for chunk in r.iter_content(chunk_size=128):
        fd.write(chunk)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
data = soup.find_all(id='csvContentDiv')[0].string
data
```

```python
clean_data = []
for row in data.split(":"):
    new_row = []
    for elem in row.split(','):
        new_row.append(elem.strip(' "'))
#     print(new_row, clean_data)
    clean_data.append(','.join(new_row))
final = '\n'.join(clean_data)
df = pd.read_csv(StringIO(final))

df
```
